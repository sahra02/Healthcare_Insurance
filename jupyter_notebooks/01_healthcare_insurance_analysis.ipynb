# Section 3: Data Quality Checks

In this section, the dataset will be checked for duplicates, missing values, incorrect data types and any other issues that may require cleaning before analysis.

```python
df.duplicated().sum()
```

This shows that there is 1 duplicate row in the dataset. The duplicate row will be removed to avoid counting the same record more than once in the analysis.

```python
df.drop_duplicates(inplace=True)
```

This removes the duplicate row from the dataset, ensuring that each customer is only counted once in the analysis.

```python
df.duplicated().sum()
```

This shows there are no duplicate rows in the dataset after removing the duplicate, as the `duplicated().sum()` method returns 0. Therefore, no further duplicate removal is required.

```python
df.shape
```

After removing the duplicate row, the dataset now contains 1,337 rows and 7 columns.

```python
df.isnull().sum()
```

This shows there are no missing values in the dataset, as the `isnull().sum()` method returns a Series of all zero values. Therefore, no missing value imputation is required.

```python
df.dtypes
```

This shows the different data types in the dataset. The `age`, `bmi`, `children` and `charges` columns are numeric, while the `sex`, `smoker` and `region` columns are object types. These data types are appropriate for the initial analysis, so no data type conversion is required at this stage.
